### ğŸ”„ Federated Learning Strategies (10 Available)

| Strategy | Description | Best Use Case |
|----------|-------------|---------------|
| **FedAvg** | Basic federated averaging | Standard FL scenarios |
| **FedAvgM** | FedAvg with server-side momentum | Improving convergence |
| **FedProx** | Proximal term for heterogeneity | Non-IID data |
| **FedNova** | Normalized averaging | Varying local epochs |
| **SCAFFOLD** | Variance reduction with control variates | Highly heterogeneous data |
| **FedAdam** | Server-side Adam optimization | Fast convergence |
| **Krum** | Byzantine-fault tolerant | Adversarial environments |
| **TrimmedMean** | Robust aggregation | Noisy updates |
| **Bulyan** | Multi-Krum + TrimmedMean | Strong Byzantine fault tolerance |
| **Multi-baseline** | Integration with Flower baselines | Advanced research |
